## I. Primera parte (40%)

**Pregunta 1.** ¿Cuál se utiliza comúnmente como medida de riesgo en los rendimientos?

A. Rendimiento esperado.

B. Inflación.

C. Desviación estándar.

D. Rendimiento medio geométrico.

Es decir, te ayuda a dictaminar de manera numérica "qué tanto" se desvían tus medidas del valor real. Por lo tanto, se utiliza como de riesgo, como si fuera una medida del "fallo" que puedes tener.

A. 4.45%

C. 5.84%

D. 6.32%

## II. Segunda parte (60%)

**Punto 1 (20%).** En el archivo `data.csv` se encuentran reportados los precios mensuales ajustados en el cierre de los activos de Microsoft, Grupo Financiero Banorte, Grupo Financiero Inbursa, Cemex y Grupo Carso.

1. Importar los precios de *Microsoft (MSFT) y de Cemex (CEMEXCPO.MX)* en un DataFrame indizado por fecha. Mostrar la gráfica de estos precios (5%).
2. Obtener un DataFrame con los rendimientos porcentuales mensuales para estos activos. Mostrar la gráfica de rendimientos (5%).
3. Reportar en un DataFrame el rendimiento esperado mensual y la volatilidad mensual para cada uno de estos activos (5%).
4. Calcular las matrices de varianza-covarianza (base mensual) y de correlación (base mensual) para estos activos (5%).

```python
# Importamos pandas y numpy y matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Importar el módulo optimize
import scipy.optimize as opt
```

```python
monthly_closes = pd.read_csv('data.csv', index_col='Date', usecols=['Date', 'MSFT', 'CEMEXCPO.MX'])
monthly_closes
```

```python
# Gráfico de histórico de precios mensuales
monthly_closes.plot(figsize=(8,6));
plt.show()
```

```python
monthly_ret = monthly_closes.pct_change().dropna()
monthly_ret
```

```python
# Podemos resumir en un DataFrame
monthly_ret_summary = pd.DataFrame(index=['Mean', 'Volatility'], columns=['MSFT', 'CEMEXCPO.MX'])
monthly_ret_summary.loc['Mean']=monthly_ret.mean()
monthly_ret_summary.loc['Volatility']=monthly_ret.std()
monthly_ret_summary
```

```python
# MATRIZ DE COVARIANZA CON RENDIMIENTOS MENSUALES
cov_matrix = monthly_ret.cov()
cov_matrix
```

```python
# MATRIZ DE CORRLECIÓN CON RENDIMIENTOS MENSUALES
corr_matrix = monthly_ret.corr()
corr_matrix
```

**Punto 2 (15%).** Usando los datos del punto anterior:

1. Construya la frontera de mínima varianza que representa las posibles combinaciones de portafolios que se pueden crear con los activos del primer punto. Reportar los rendimientos esperados y las volatilidades de estos portafolios en un DataFrame (8%).
2. Encontrar los pesos del portafolio de mínima varianza. ¿Cuáles son el rendimiento esperado y volatilidad del portafolio de mínima varianza? (7%).

```python
# Vector de w variando entre 0 y 1 con n pasos
wp = np.linspace(0, 1, 30)
# Rendimientos esperados individuales
#Activo 1: RU  ;  Actiovo 2: Japon
E1 = monthly_ret_summary['MSFT']['Mean']
E2 = monthly_ret_summary['CEMEXCPO.MX']['Mean']

# Volatilidades individuales
s1 = monthly_ret_summary['MSFT']['Volatility']
s2 = monthly_ret_summary['CEMEXCPO.MX']['Volatility']

#Correlacion
r12 = corr_matrix['MSFT']['CEMEXCPO.MX']

# Crear un DataFrame cuyas columnas sean rendimiento
# y volatilidad del portafolio para cada una de las w
# generadas

# w = MSFT  
#(1-w) = CEMEX

portafolios1=pd.DataFrame(columns=['Mean','Volatility'], index=wp)
portafolios1.index.name= 'w'
portafolios1.Mean = wp*E1+(1-wp)*E2
portafolios1.Volatility=np.sqrt((wp*s1)**2+((1-wp)*s2)**2+2*wp*(1-wp)*r12*s1*s2)
portafolios1.round(4)
```

```python
wp1_minvar1 =  (s2**2-r12*s1*s2)/(s1**2+s2**2-2*r12*s1*s2)
wp1_minvar1 #MSFT
##Porcentaje de portafolio que se invertirá en la ponderación w (MSFT en este caso) 1
```

```python
# Función objetivo
def var(wp1, s1, s2, s12):
    return (wp1*s1)**2 + ((1-wp1)*s2)**2 + 2*wp1*(1-wp1)*s12

# Dato inicial
w0 = 0
# Volatilidades individuales y covarianza
s1 = monthly_ret_summary['MSFT']['Volatility']
s2 = monthly_ret_summary['CEMEXCPO.MX']['Volatility']
s12 = corr_matrix['MSFT']['CEMEXCPO.MX']*s1*s2
# Cota de w
bnd = (0,1)

# Solución
min_var1 = opt.minimize(var, w0, args=(s1,s2,s12), bounds=(bnd,))
min_var1['x']
```

```python
#Volatilidad de Port. Mínima Varianza 1
wp1 = wp1_minvar1 
Vol1=np.sqrt((wp1*s1)**2+((1-wp1)*s2)**2+2*wp1*(1-wp1)*r12*s1*s2)
Vol1
```

```python
#Rendimiento de Port. Mínima Varianza 1
E1 = monthly_ret_summary['MSFT']['Mean']
E2 = monthly_ret_summary['CEMEXCPO.MX']['Mean']
Er_p1 = wp1*E1+(1-wp1)*E2
Er_p1
```

**Punto 3 (15%).** En este punto se usarán los mismos activos con sus rendimientos esperados y volatilidades, solo que supondremos que el coeficiente de correlación entre sus rendimientos se supondrá de 0.5.

2. Encontrar los pesos del portafolio de mínima varianza. ¿Cuáles son el rendimiento esperado y volatilidad del portafolio de mínima varianza? (7%).

```python
# Vector de w variando entre 0 y 1 con n pasos
wp = np.linspace(0, 1, 30)
# Rendimientos esperados individuales
#Activo 1: RU  ;  Actiovo 2: Japon
E1 = monthly_ret_summary['MSFT']['Mean']
E2 = monthly_ret_summary['CEMEXCPO.MX']['Mean']

# Volatilidades individuales
s1 = monthly_ret_summary['MSFT']['Volatility']
s2 = monthly_ret_summary['CEMEXCPO.MX']['Volatility']

#Correlacion
r12 = 0.5

# Crear un DataFrame cuyas columnas sean rendimiento
# y volatilidad del portafolio para cada una de las w
# generadas

# w = MSFT  
#(1-w) = CEMEX

portafolios2=pd.DataFrame(columns=['Mean','Volatility'], index=wp)
portafolios2.index.name= 'w'
portafolios2.Mean = wp*E1+(1-wp)*E2
portafolios2.Volatility=np.sqrt((wp*s1)**2+((1-wp)*s2)**2+2*wp*(1-wp)*r12*s1*s2)
portafolios2.round(4)
```

```python
wp2_minvar2 =  (s2**2-r12*s1*s2)/(s1**2+s2**2-2*r12*s1*s2)
wp2_minvar2 #MSFT
##Porcentaje de portafolio que se invertirá en la ponderación w (MSFT en este caso) 2
```

```python
# Función objetivo
def var(wp2, s1, s2, s12):
    return (wp2*s1)**2 + ((1-wp2)*s2)**2 + 2*wp2*(1-wp2)*s12

# Dato inicial
w0 = 0
# Volatilidades individuales y covarianza
s1 = monthly_ret_summary['MSFT']['Volatility']
s2 = monthly_ret_summary['CEMEXCPO.MX']['Volatility']
s12 = 0.5*s1*s2
# Cota de w
bnd = (0,1)

# Solución
min_var2 = opt.minimize(var, w0, args=(s1,s2,s12), bounds=(bnd,))
min_var2['x']
```

```python
#Volatilidad de Port. Mínima Varianza 2
wp2 = wp2_minvar2 
Vol2=np.sqrt((wp2*s1)**2+((1-wp2)*s2)**2+2*wp2*(1-wp2)*r12*s1*s2)
Vol2
```

```python
#Rendimiento de Port. Mínima Varianza 2
E1 = monthly_ret_summary['MSFT']['Mean']
E2 = monthly_ret_summary['CEMEXCPO.MX']['Mean']
Er_p2 = wp2*E1+(1-wp2)*E2
Er_p2
```
